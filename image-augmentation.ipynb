```python
import os.path
from os import path
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import numpy as np
```

```python
from zipfile import ZipFile

directory = "./drive/MyDrive/Colab Notebooks/NFT-art/"      # directory to where image is 
image = load_img(os.path.join(directory, "sample_gen.jpg"))     # input image file

image = img_to_array(image)
image = np.expand_dims(image, axis=0)
aug = ImageDataGenerator()

if path.exists('./drive/MyDrive/Colab Notebooks/NFT-art/generated_images') == False:
  os.mkdir('./drive/MyDrive/Colab Notebooks/NFT-art/generated_images')

new_directory = "./drive/MyDrive/Colab Notebooks/NFT-art/generated_images"      # to save augmented images

count = 0
imageGen = aug.flow(image, batch_size=1, save_to_dir=new_directory,
                    save_prefix="image", save_format="jpg")

total = 50      # Number of copies to generate

for i in imageGen:
    count += 1
    if count == total:
        break

# Zipping generated images

!zip -r "./drive/MyDrive/Colab Notebooks/NFT-art/generated_images.zip"  "./drive/MyDrive/Colab Notebooks/NFT-art/generated_images"

# Generated images file download 

from google.colab import files #To download
#files.download('./drive/MyDrive/Colab Notebooks/ai-igbo-caps-pixelate/generated_images.zip')
```
